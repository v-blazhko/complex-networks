```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/dataset.csv', header=0, names=['id', 'name', 'lat', 'lon', 'locality', 'region', 'neighbourhood'])
data = [df['lat'], df['lon']]
headers = ['lat', 'lon']
df = pd.concat(data, axis=1, keys=headers)
df.head()
```

```python
img = plt.imread('map.png')
bbox = (20.410, 30.410, 59.800, 69.910)
```

```python
fig, ax = plt.subplots(figsize = (10,24))
ax.scatter(df.lon, df.lat, zorder=1, alpha= 0.1, c='b', s=5)
ax.set_title('Plotting Spatial Data')
ax.set_xlim(bbox[0],bbox[1])
ax.set_ylim(bbox[2],bbox[3])
ax.imshow(img, zorder=0, extent = bbox, aspect= 'equal')
```
