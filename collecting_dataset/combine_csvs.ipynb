```python
import pandas as pd
```

```python
df_1 = pd.read_csv('dataset/raw/veronika_all.csv', header=0, names=['ind', 'id', 'name', 'lat', 'lon', 'locality', 'region', 'neighbourhood'])
df_1 = df_1.drop(['ind'], axis=1)
df_1.head()
```

```python
df_2 = pd.read_csv('dataset/raw/ashish_1.csv', header=0, names=['id', 'lat', 'lon', 'name', 'locality', 'neighbourhood', 'region'])
df_2.head()
```

```python
df_3 = pd.read_csv('dataset/raw/ashish_2.csv', header=0, names=['id', 'lat', 'lon', 'name', 'locality', 'neighbourhood', 'region'])
df_3.head()
```

```python
df_4 = pd.read_csv('dataset/raw/ashish_3.csv', header=0, names=['id', 'lat', 'lon', 'name', 'locality', 'neighbourhood', 'region'])
df_4.head()
```

```python
df = pd.concat([df_1, df_2, df_3, df_4])
```

```python
duplicateRowsDF = df[df.duplicated()]
len(duplicateRowsDF)
```

```python
df = df.drop_duplicates(subset='id').reset_index(drop=True)
len(df)
```

```python
df.to_csv('dataset/dataset.csv', index=False)
```
