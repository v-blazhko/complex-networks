```python
import pandas as pd
from collections import Counter
from tqdm import tqdm
from vincenty import vincenty
import geopandas
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
import numpy as np
from sklearn.cluster import KMeans
from numpy import vstack,array
from numpy.random import rand
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans2, whiten, kmeans,vq
from itertools import combinations
import networkx as nx
```

```python
df = pd.read_csv('../dataset/dataset_with_users_and_clusters.csv')
df.head()
```

```python
sizes = df.groupby(['cluster_id']).size().reset_index().rename(columns={0:'count'})['count'].values
```

```python
clusters = pd.read_csv('../dataset/clusters.csv')
clusters.index.name = 'cluster_id'
clusters['size'] = sizes
clusters.to_csv('../dataset/clusters.csv', index=0)
clusters.head()
```

```python
labels = []
for c in tqdm(clusters.iterrows()):
    localities = list(df[(df['cluster_id']==c[0])].locality)
    regions = list(df[(df['cluster_id']==c[0])].region)
    l = max(set(localities), key=localities.count)
    r = max(set(regions), key=regions.count)
    label = str(l) + ', ' + str(r)
    labels.append(label)
```

```python
clusters['label'] = labels
clusters.to_csv('../dataset/clusters.csv', index=0)
clusters.head()
```
