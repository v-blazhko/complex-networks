```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('dataset/distances.csv')
```

```python
data.head()
```

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
nodes=[]
for i in range(0,500):
    nodes.append(i)
```

```python
G.add_nodes_from(nodes)
```

```python
source=data["source"]
```

```python
destination=data["destination"]
```

```python
weights=data["v_distance"]
```

```python
for i in range(0,len(source)):
    G.add_weighted_edges_from([(source[i],destination[i],weights[i])])
```

```python
G.nodes()
```

```python
nx.draw(G, node_size=200, with_labels=True, node_color='w', edgecolors='k')
```

```python
nx.write_gexf(G, "test.gexf")
```

```python
T=nx.minimum_spanning_tree(G)
```

```python
nx.write_gexf(T,"min_test.gexf")
```

```python
nx.draw(T, node_size=200, with_labels=True, node_color='b', edgecolors='k')
```
